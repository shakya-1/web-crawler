```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
import time
from newspaper3k import Article 
```

```python
articles=open("articles.txt",'w')
```

```python
def get_news_dailymail():
    
    # url definition
    url = "https://www.livemint.com/topic/international"
    
    # Request
    r1 = requests.get(url)
    r1.status_code

    # We'll save in coverpage the cover page content
    coverpage = r1.content

    # Soup creation
    soup1 = BeautifulSoup(coverpage, 'html.parser')

    # News identification
    coverpage_news = soup1.find_all('h2', class_='headline')
    len(coverpage_news)
    
    number_of_articles = 5

    # Empty lists for content, links and titles
    news_contents = []
    list_links = []
    list_titles = []

    for n in np.arange(0, number_of_articles):

        # Getting the link of the article
        link = 'https://www.livemint.com' + coverpage_news[n].find('a')['href']
        list_links.append(link)
     

        # Getting the title
        title = coverpage_news[n].find('a').get_text()
        list_titles.append(title)
        
   
    for i in list_links:
            print(i)
            t_article = Article(i, language="en") # en for English 
            #To download the article 
            t_article.download() 
            #To parse the article 
            t_article.parse() 
            #To perform natural language processing ie..nlp 
            t_article.nlp()
            print(t_article.title) 
            print(" ") 
            #To extract text 
            print(t_article.text) 
            #writing the articles to the text file
            articles.write(t_article.text)
            print(" ") 
            #To extract summary 
            print(t_article.summary) 
    articles.close()


    # df_show_info
    df_show_info = pd.DataFrame(
        {'Article Title': list_titles,
         'Article Link': list_links,
         'Newspaper': 'Daily Mail'})
    
    return (df_show_info)

```

```python
f=open("articles.txt",'r')
text=f.read()
#print(text)
text=text.replace('Share Via','sharevia ')
text=text.replace('Updated:','updated ')
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
sentences=sent_tokenize(text)
#sentences
```

```python
fi=open("rahul.txt",'w')
lis=[]
undesired_words=['sharevia','Updated']
for sen in sentences:
    lis.append(sen)
for i in range(len(lis)):
    wo=word_tokenize(lis[i])
    if(wo[0] not in undesired_words):
        fi.write(lis[i]+"\n") 
```

```python
fi=open("rahul.txt",'r+')
s=(fi.read())
s=sent_tokenize(s)
fi.truncate(0)
fi.close()
fi=open("rahul.txt","w")
for i in range(len(s)):
    fi.write('{}'.format(i)+"\t")
    fi.write(str(s[i])+"\n")
fi.close()
```

```python
import csv
with open("mycsv.csv",'w') as f:
    fieldnames=['ID','Sentences']
    thewriter=csv.DictWriter(f,fieldnames=fieldnames)
    thewriter.writeheader()
    for i in range(len(s)):
        thewriter.writerow({'ID':i,'Sentences':s[i]})
```
